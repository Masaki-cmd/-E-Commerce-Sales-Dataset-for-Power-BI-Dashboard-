### KGI: 利益率の向上 (Improve Profit Margin)

### 提案されたKPI:
1.  **総利益 (Total Profit)**: `profit_per_order` 列の合計。
2.  **総売上高 (Total Sales)**: `sales_per_order` 列の合計。
3.  **平均注文単価 (Average Order Value - AOV)**: 総売上高 / (ユニークな`order_id`の数)。
4.  **注文数 (Number of Orders)**: ユニークな`order_id`の数。
5.  **平均割引率 (Average Discount Rate)**: `order_item_discount` / `sales_per_order` の平均。
6.  **利益率 (Profit Margin)**: (総利益 / 総売上高) * 100。

### KPIツリーの概要:
KGI「利益率の向上」は、**利益率**によって測定され、これは**総利益**と**総売上高**に分解されます。

-   **総売上高**は、**平均注文単価 (AOV)** と**注文数**によって構成されます。

このツリーは、利益率を向上させるための具体的なレバー（総利益の増加、総売上高の増加、平均注文単価の向上、注文数の増加、割引率の最適化）を視覚的に示します。
